```python
from pmdarima import auto_arima
import pandas as pd
```

```python
FILE_DATOS = 'lagoon_hourly_filled.csv'
parametro = 'temperatura'
output=parametro
inputs = [
          'fecha', 'temperatura',
          'ambiente', 'nivel'
          ]
usecols = inputs.copy()

datos = pd.read_csv(FILE_DATOS,
                    sep=',',
                    usecols = usecols)

# Preparaci√≥n del dato
# ==============================================================================
datos['fecha'] = pd.to_datetime(datos['fecha'], format='%d/%m/%Y %H:%M')
datos = datos.set_index('fecha')
datos = datos.rename(columns={'x': 'y'})
datos = datos.asfreq('H')
datos = datos.sort_index()

datos.info()

datos.head()
```

```python
modelo_auto=auto_arima(datos['temperatura'],
                       start_p=0,d=1,start_q=0,
         max_p=10,max_d=3,max_q=10, m=24, seasonal=False,
         error_action='warn',trace=True,
         supress_warnings=True,stepwise=True,
         random_state=20,n_fits=100)
print(modelo_auto)
```

```python
modelo_auto=auto_arima(datos['temperatura'],
                       start_p=0,d=1,start_q=0,
         max_p=10,max_d=3,max_q=10, m=24, seasonal=True,
         error_action='warn',trace=True,
         supress_warnings=True,stepwise=True,
         random_state=20,n_fits=100)
print(modelo_auto)
```
